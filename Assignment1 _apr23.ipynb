```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#read the excel data set named iris
iris_data=pd.read_excel(r'C:\Users\DELL\Downloads\iris.xls')
```

```python
#head() shows the first five entries of the data set iris
iris_data.head()
```

```python
#Here it displays the columns of the data set iris
iris_data.columns
```

```python
#Calculated the mean of each column(SL,SW,PL,PW) in the data set by using the function mean()
iris_data.mean()
```

```python
#Checking the presence of null values in the data and founded that there is no null values
iris_data.isna().sum()
```

```python
#Visualization of the classification of the iris flower and founded that there are 50 counts of each species
sns.countplot(x='Classification',data=iris_data)
plt.title("Count of Classifications")
plt.show()
```

There are 50 entries of each Iris-Setosa,Iris-Versicolor and Iris-Virginica species

```python
#finding which classification have lengthy sepal
plt.bar(iris_data['Classification'],iris_data['SL'])
plt.ylabel("Sepal Length")
plt.xlabel("Classification")
plt.title("Sepal length of classification")
plt.show()
```

From above graph, found that iris virginica have large sepal length.

```python
#Comparing sepal length and sepal width of iris classifications
sns.scatterplot(x='SL',y='SW',hue="Classification",data=iris_data)
plt.legend(bbox_to_anchor=(1,1),loc=2)
plt.title("Comparison of SL and SW")
plt.show()
```

1.Iris-Setosa have smaller sepal length but larger sepal width

2.Iris-Versicolor in the middle of the other classifation in terms of sepal length and sepal width

3.Iris-Verginica have larger sepal length but smaller sepal width

```python
#Comparison of petal length and petal width of the classifications
sns.scatterplot(x='PL',y='PW',hue="Classification",data=iris_data)
plt.legend(bbox_to_anchor=(1,1),loc=2)
plt.title("Comparison of PL and PW")
plt.show()
```

1.Iris-Setosa have smaller petal length and petal width

2.Iris-Versicolor lies in the middle of other classification in terms of petal length and petal width

3.Iris-virginica have larger petal length and petal width
